```python
import ast
```

```python
def str2ls(obj):
    ls=[]
    for i in ast.literal_eval(obj):
        ls.append(i['name'])
    return ls
```

```python
mov.iloc[0]['crew']
```

```python
def str2ls5(obj):
    ls=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=5:
            ls.append(i['name'])
            count+=1
        else:
            break
    return ls
```

```python
def directname(obj):
    ls=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':   
            ls.append(i['name'])
        else:
            break
    return ls
```

```python
mov['genres']=mov['genres'].apply(str2ls)
```

```python
mov['keywords']=mov['keywords'].apply(str2ls)
```

```python
mov['cast']=mov['cast'].apply(str2ls5)
```

```python
mov['crew']=mov['crew'].apply(directname)
```

```python
mov
```

```python
mov['overview']=mov['overview'].apply(lambda x:x.split())
```

```python
mov
```

```python
mov['genres'] = mov['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
mov['keywords'] = mov['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
mov['cast'] = mov['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
mov['crew'] = mov['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
mov
```

```python
mov['tags'] = mov['overview'] + mov['genres'] + mov['keywords'] + mov['cast'] + mov['crew']
```

```python
mov
```

```python
mov=mov[['title','id','tags']]
```

```python
mov
```

```python
mov['tags']=mov['tags'].apply(lambda x: ' '.join(x))
```

```python
mov
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
mov['tags']=mov['tags'].apply(stem)
```

```python
mov
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5500,stop_words='english')
```

```python
vectors=cv.fit_transform(mov['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
close=cosine_similarity(vectors)
```

```python
close.shape
```

```python
close[0]
```

```python
mov[mov['title']=='Avatar'].index[0]
```

```python
def movieRecommend(movie):
    movie_index=mov[mov['title']==movie].index[0]
    differences=close(movie_index)
    movie_lists=sorted(list(enumerate(differences)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_lists:
        print(mov.iloc[i[0]].title)
```

```python
movieRecommend('Batman begins')
```
